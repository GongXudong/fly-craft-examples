```python
import numpy as np
from stable_baselines3.common.running_mean_std import RunningMeanStd
```

```python
arr1 = np.arange(100) * (-0.01)
arr1 = arr1[::-1].reshape([-1, 1])

arr2 = np.arange(10) * (-0.1)
arr2 = arr2[::-1].reshape([-1, 1])
arr2 = arr1[::-1]

arr1, arr2
```

```python
def test_var(arr: np.ndarray):
    R = 0.
    gamma = 0.98
    rms = RunningMeanStd()
    for item in arr:
        R = R * gamma + item
        rms.update(R)
    
    return rms.var
```

```python
test_var(arr1), test_var(arr2)
```

```python
tmp_arr1 = arr1 / test_var(arr1)
tmp_arr2 = arr2 / test_var(arr2)
tmp_arr1, tmp_arr2
```

```python
arr1.sum(), tmp_arr1.sum(), arr2.sum(), tmp_arr2.sum()
```
