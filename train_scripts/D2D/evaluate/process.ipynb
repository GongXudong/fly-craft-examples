```python
from pathlib import Path
import numpy as np
import pandas as pd
from flycraft.utils_common.geometry_utils import angle_of_2_velocity
PROJECT_ROOT_DIR = Path().absolute().parent.parent.parent
PROJECT_ROOT_DIR
```

```python
def process_one_file(file_path: Path):
    df = pd.read_csv(file_path)
    df["v_error"] = np.abs(df["v"] - df["achieved v"])
    df["angle_error"] = df.apply(lambda row: np.abs(angle_of_2_velocity(row["v"], row["mu"], row["chi"], row["achieved v"], row["achieved mu"], row["achieved chi"])), axis=1)
    return df
```

```python
df = process_one_file(PROJECT_ROOT_DIR / "train_scripts" / "D2D" / "evaluate" / "cache" / "sac_hard_b_2_1e6_1.csv")
```
